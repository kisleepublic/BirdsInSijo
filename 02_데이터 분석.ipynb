```python
df_bird = df_bird.fillna('')
```

```python
poem_list = []
for index, row in df_bird.iterrows():
    poem = []
    poem.append(row['번호'])
    tokens = row['초장'].split(',')
    temp = []
    for tok in tokens:
        if tok:
            temp.append(tok)
    poem.append(temp)
    
    tokens = row['중장'].split(',')
    temp = []
    for tok in tokens:
        if tok:
            temp.append(tok)
    poem.append(temp)
    
    tokens = row['종장'].split(',')
    temp = []
    for tok in tokens:
        if tok:
            temp.append(tok)
    poem.append(temp)
  
    poem_list.append(poem)
```

```python
poem_list
```

```python
def unify_bird_name(name):
    for key, value in name_dic.items():
        if name in value:
            return key
    print('찾지 못함', name)
    return '해당없음'    
```

```python
#새들을 대표 명칭으로 통일
for i in range(len(poem_list)):
    #번호 0, 초장 1, 중장 2, 종장 3
    for j in range(len(poem_list[i][1])): #초장 탐색
        poem_list[i][1][j] = unify_bird_name(poem_list[i][1][j])
    for j in range(len(poem_list[i][2])): #중장 탐색
        poem_list[i][2][j] = unify_bird_name(poem_list[i][2][j])
    for j in range(len(poem_list[i][3])): #종장 탐색
        poem_list[i][3][j] = unify_bird_name(poem_list[i][3][j])       
```

```python
poem_list
```

```python
bird_list = []
for row in poem_list:
    birds = []
    for bird in row[1]:
        birds.append(bird)
    for bird in row[2]:
        birds.append(bird)
    for bird in row[3]:
        birds.append(bird)
    bird_list.append(list(set(birds)))
```

```python
bird_list
```

```python
def index_to_poem_num(num):
    return poem_list[num][0]
```

```python
#여러 새가 동시에 출현하는 시조 번호 조회
for i in range(len(bird_list)):
    if len(bird_list[i])==3: #동시출현 새의 수 조정 파라미터
        print(index_to_poem_num(i), ': ', end='')
        for b in bird_list[i]:
            print(b, end=' ')
        print()
```

```python
freq_dic = {}
```

```python
#각 새의 출현 빈도
for birds in bird_list:
    for bird in birds:
        freq_dic[bird] = freq_dic.get(bird, 0)+1
```

```python
freq_list = []
for key, value in freq_dic.items():
    freq_list.append((value,key))
freq_list.sort(reverse=True)
```

```python
bird_pair=[]
for birds in bird_list:
    if len(birds) > 1:
        pairs = list(itertools.combinations(birds, 2)) #동시출현 대상에 대해 두 쌍으로 조합 생성
        bird_pair += pairs        
```

```python
#특정 새가 등장하는 시조 번호 조회
cnt = 0
for i in range(len(bird_list)):
    if '기러기' in (bird_list[i]):
        print(index_to_poem_num(i), end=' ')
        cnt = cnt + 1
#print(cnt)
```

```python
#동시출현 조합에 대한 빈도수

# 각 시조의 새 리스트를 정렬하여 동일한 새 집합을 동일하게 표현
sorted_sijo_list = [sorted(sijo) for sijo in bird_list]

# 각 새 집합이 몇 번 등장하는지 카운트
sijo_count = {}
for sijo in sorted_sijo_list:
    if len(sijo) != 2 : #동시출현 새의 수
        continue

    # 리스트를 튜플로 변환하여 딕셔너리의 키로 사용
    sijo_tuple = tuple(sijo)
    if sijo_tuple in sijo_count:
        sijo_count[sijo_tuple] += 1
    else:
        sijo_count[sijo_tuple] = 1

# 새 집합이 등장하는 횟수에 따라 내림차순으로 정렬
sorted_sijo_count = sorted(sijo_count.items(), key=lambda item: item[1], reverse=True)

for birds, count in sorted_sijo_count:
    # 새의 집합이 2번 이상 등장하는 경우에만 출력
    if count >= 2:
        print(f"{list(birds)}: {count}")
```

```python
#동시출현으로 나타난 적이 없는 새

# 모든 새를 집합으로 모음
all_birds = set(bird for sijo in bird_list for bird in sijo)

# 다른 새와 함께 묶인 새를 모음
paired_birds = set(bird for sijo in bird_list for bird in sijo if len(sijo) > 1)

# 다른 새와 함께 묶인 적이 없는 새를 찾음
unpaired_birds = all_birds - paired_birds

print("다른 새와 함께 묶인 적이 없는 새:")
for bird in unpaired_birds:
    print(bird)
```
